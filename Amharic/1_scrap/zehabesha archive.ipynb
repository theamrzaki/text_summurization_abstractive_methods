```
from bs4 import BeautifulSoup
import requests
import pandas as pd

def get_number_of_pages(link):
  response = requests.get(link)
  data = response.text
  soup = BeautifulSoup(data, 'lxml')
  number_of_pages = soup.find("div", {"class": "page-meta"})
  number_of_pages = int(number_of_pages.text.split(" ")[-1].strip())
  return number_of_pages

def get_hrefs(link):
  response = requests.get(link)
  data = response.text
  soup = BeautifulSoup(data, 'lxml')
  h2s = soup.find_all("h2", {"class": "entry-title e-t"})
  hrefs = []
  for h in h2s:
    hrefs.append(h.find_all("a")[0].attrs["href"])
  return hrefs

def get_content(link):
  response = requests.get(link)
  data = response.text
  soup = BeautifulSoup(data, 'lxml')
  title = soup.find("h1", {"class": "headline entry-title e-t m-b__20"}).text
  content = soup.find("div", {"class": "entry-content"})
  content = " ".join([c.text for c in content.find_all("p")])
  return title,content
```

```
year_start = 2011
year_end   = 2020

month_start = 1
month_end   = 12

i = 0
save_period = 100
output = []

for year in range(year_start,year_end+1):
  for month in range(month_start,month_end+1):
    try:
      link = "https://www.zehabesha.com/amharic/"+str(year)+"/"+str(month)+"/"
      number_of_pages = get_number_of_pages(link)
      for num in range(1,number_of_pages+1):
        num_link = link +"page"+"/"+str(num)+"/"
        hrefs = get_hrefs(num_link)
        for h in hrefs:
          title,content = get_content(h)
          row = {}
          row["title"] = title
          row["content"] = content
          row["link"] = str(h)
          row["id"] = str(year) + "#"+ str(month) + "#"+ str(num) 
          output.append(row)
          i+=1
          print("----" + str(i))
          if i % save_period == 0 and len(output) > 0 and i != 0:
            print("saving")
            pd.DataFrame(output).to_csv("/content/drive/My Drive/Amhari/news_zehabesha_archive.csv", mode='a',index = False,encoding='utf-8')
            output = []
    except Exception as e :
      print("##" + str(e) + "##" +  link )
```
